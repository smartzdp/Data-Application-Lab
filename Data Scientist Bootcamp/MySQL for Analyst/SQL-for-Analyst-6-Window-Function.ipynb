```python
%%sql

select
    *
from
    simple_sales
order by country, year, product
;
```

```python
%%sql

select
    country, 
    sum(profit) as country_profit
from
    simple_sales
group by country
order by country
;
```

```python
%%sql

select
    year, country, product, profit, 
    sum(profit) over(partition by country) as country_profit
from
    simple_sales
order by country, year, product
;
```

```python
%%sql

select
    year, country, product, profit, 
    row_number() over(partition by country order by profit) as row_num1, 
    row_number() over(partition by year order by profit) as row_num2
from
    simple_sales
;
```

```python
%%sql

select
    concat(first_name, ' ', last_name) as full_name, 
    salary, 
    avg(salary) over() as avg_salary, 
    sum(salary) over() as sum_salary
from
    employees
limit 10
;
```

```python
%%sql

select
    concat(first_name, ' ', last_name) as full_name, 
    salary, 
    avg(salary) over(partition by department_id) as avg_department_salary, 
    sum(salary) over() as sum_salary
from
    employees
limit 10
;
```

```python
%%sql

select
    employee_id, 
    concat(first_name, ' ', last_name) as full_name, 
    row_number() over(order by salary desc) as row_num, 
    rank() over(order by salary desc) as rank_num, 
    dense_rank() over(order by salary desc) as dense_rank_num, 
    salary
from
    employees
limit 10
;
```

```python
%%sql

select
    *
from
    (select
        employee_id, 
        concat(first_name, ' ', last_name) as full_name, 
        row_number() over(order by salary desc) as row_num, 
        rank() over(order by salary desc) as rank_num, 
        dense_rank() over(order by salary desc) as dense_rank_num, 
        salary
    from
        employees) as e
where
    e.dense_rank_num <= 3
;
```

```python
%%sql

select
    symbol, date, price, 
    avg(price) over (partition by symbol order by date rows unbounded preceding) as price_ma
from
    ts
;
```

```python
%%sql

select
    symbol, date, volume, 
    sum(volume) over (partition by symbol order by date rows unbounded preceding) as volume_ms
from
    ts
;
```

```python
%%sql

select
    symbol, date, price, 
    avg(price) over (partition by symbol order by date rows between 3 preceding and 3 following) as price_cma_7r
from
    ts
;
```

```python
%%sql

select
    symbol, date, price, 
    avg(price) over (partition by symbol order by date range between interval 6 day preceding and current row) as price_ma_7d, 
    avg(price) over (partition by symbol order by date range between interval 1 week preceding and current row) as price_ma_1w, 
    avg(price) over (partition by symbol order by date range between interval 1 month preceding and current row) as price_ma_1m
from
    ts
;
```

```python
%%sql

select
    symbol, date, price, 
    avg(price) over (partition by symbol order by date range between interval 3 day preceding and interval 3 day following) as price_cma_7d
from
    ts
;
```

```python
%%sql

select
    row_number() over() as row_num, 
    employee_id, 
    concat(first_name, ' ', last_name) as full_name, 
    first_value(salary) over() as 'first', 
    last_value(salary) over() as 'last', 
    nth_value(salary, 2) over() as 'second', 
    nth_value(salary, 4) over() as 'fourth', 
    lag(salary, 1, 0) over() as 'lag_salary', 
    salary, 
    lead(salary, 1, 0) over() as 'lead_salary'
from
    employees
limit 10
;
```
